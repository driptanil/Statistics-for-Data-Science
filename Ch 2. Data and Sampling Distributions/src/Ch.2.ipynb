```python
import pandas as pd
```

```python
player_data_import = pd.read_csv('../data/players.csv')
player_data_import
```

```python
def heightInches(height):
    my_list = height.split('-')
    if len(my_list) == 1:
        return int (my_list[0])
    else:
        return int (my_list[0]) * 12 + int (my_list[1])
```

```python
player_data_import['height_inches'] = player_data_import['height'].apply(heightInches)
```

```python
player_data_import
```

```python
import matplotlib.pyplot as plt

sample_amount = range(10, 100, 20)

for x in sample_amount:
    samples = []
    for i in range(1000):
        samples.append(player_data_import['height_inches'].sample(x).mean())
    plt.hist(samples)

plt.legend([str(z) + " samples" for z in sample_amount])
plt.title("Central Limit Theorem")
```
